```python
import mysql.connector #import library untuk menghubungkan ke mysql

# buat koneksi ke database
db = mysql.connector.connect(
  host="localhost",
  user="root",
  password=""
)

# buat kursor
crs = db.cursor()

# buat database db_sales_V3922026
crs.execute("CREATE DATABASE db_sales_V3922026")

print("Database berhasil dibuat")

# # menggunakan database db_sales_V3922026
crs.execute("USE db_sales_V3922026")

print("Database berhasil digunakan")

# membuat table data stok_barang dengan query create table
crs.execute("CREATE TABLE stok_barang (\
  id_barang VARCHAR(10) PRIMARY KEY, \
  nama_barang VARCHAR(30) NOT NULL, \
  harga_barang int NOT NULL, \
  stok_awal int NOT NULL, \
  barang_masuk int NOT NULL, \
  barang_keluar int NOT NULL, \
  stok_akhir int NOT NULL\
)")

print("Table berhasil dibuat")
# menutup koneksi database
db.close()
```

```python
# Membuat function insert
def insert_data():
    pil = 1 # menginsialisasikan nilai awal variabel pil dengan 1
    while pil==1: # membuat perulangan while apabila pil==1 maka kode dibawah ini akan dijalankan 
        print("---------INSERT DATA BARANG---------")
        idbarang = input("Masukkan ID Barang: ") # meminta input id barang
        crs.execute(f"SELECT * FROM stok_barang WHERE id_barang='{idbarang}'") # mengambil data dari tabel "stok_barang" dimana id_barang sesuai dengan input user
        result = crs.fetchone() # menyimpan hasil query dalam variabel result
        if result is not None: # jika hasil query tidak kosong, maka ID Barang sudah ada di tabel
            print("ID Barang sudah ada. Mohon masukkan ID Barang yang berbeda.")
            continue # lanjutkan ke input kembali
        nama = input("Masukkan Nama Barang:") # meminta input nama barang
        harga = int(input("Masukkan Harga Barang:")) # meminta input Harga Barang
        stok1 = int(input("Masukkan Stok Awal Barang:")) # meminta input Stok Awal Barang
        barangIn = int(input("Masukkan Jumlah Barang Masuk:")) # meminta input Jumlah Barang Masuk
        barangOut = int(input("Masukkan Jumlah Barang Keluar:"))# meminta input Jumlah Barang Keluar
        stok2 = stok1 + barangIn - barangOut # menghitung stok akhir barang
        print(f"Stok Akhir Barang adalah {stok2}")
        crs.execute(f"INSERT INTO stok_barang (id_barang, nama_barang, harga_barang, stok_awal, barang_masuk, barang_keluar, stok_akhir) \
        VALUES ('{idbarang}', '{nama}', {harga}, {stok1}, {barangIn}, {barangOut}, {stok2})") # memasukkan data ke tabel "stok_barang" dengan query INSERT
        db.commit() # menyimpan perubahan pada database
        print("Insert Data Berhasil")
        print("-----------------------------------")
        pil = int(input("Masukkan data lagi? 1.Ya \n 0. Tidak: ")) # menanyakan apakah ingin memasukkan data lagi atau tidak
```

```python
# Membuat function utntuk menampilkan data barang dari database
def show_data():
    print("---------MENAMPILKAN DATA BARANG---------")
    crs.execute("SELECT * FROM stok_barang") # meneksekusi query select * untuk menapilkan semua data pada table stok_barang
    result = crs.fetchall() # menggunakan method fetchall untuk mengambil semua baris data dari hasil query database. 
    for row in result: # menampilkan dengan perulangan for untuk setiap row data dari table
        idbarang = row[0] # menyimpan nilai pada setiap kolom yang diambil menggunakan indeks array ke dalam variable
        nama = row[1]
        stok1 = row[2]
        barangIn = row[3]
        barangOut = row[4]
        stok2 = row[5]
        # menampilkan data 
        print(f"ID Barang: {idbarang}\nNama Barang: {nama}\
        \nStok Awal: {stok1}\nJumlah Barang Masuk: {barangIn}\nJumlah Barang Keluar: {barangOut}\nStok Akhir: {stok2}\n")
    print("----------------------------------")
    print("Menampilkan Data Berhasil")
    print("-----------------------------------")
```

```python
# membuat function untuk mengupdate data barang

def update_data():
    pil = 1 # menginsialisasikan nilai awal variabel pil dengan 1
    while pil==1: # membuat perulangan while apabila pil==1 maka kode dibawah ini akan dijalankan 
        print("---------UPDATE DATA BARANG---------")
        crs.execute("SELECT * FROM stok_barang") # mengeksekui query untuk mengambil data barang
        result = crs.fetchall()
        for row in result: # menampilkan data barang menggunnakan perulanagan for
            idbarang = row[0] # hanya menampilkan id dan nama barang saja
            nama = row[1]
            print("Data Barang")
            print(f"ID Barang: {idbarang}\nNama Barang: {nama}\n")
        idbarang = input("Masukkan ID Barang yang ingin diupdate: ") # meminta input id barang
        barangIn= input("Masukkan jumlah barang masuk: ")  # meminta input Jumlah Barang Masuk
        barangOut = input("Masukkan jumlah barang keluar: ") # meminta input Jumlah Barang Keluar
        # mengeksekusi query memilih mengambil dari table t=stok barnag berdasrkan id barang
        crs.execute(f"SELECT stok_akhir FROM stok_barang WHERE id_barang = '{idbarang}'") 
        result = crs.fetchone() # menyimpan hasil query dalam variabel result
        if result is None: # membuat pengkondisian apabila query pengambilan data barang berdasarkan id tidak ditemukan maka  akan muncul pesan error 
            print("ID Barang tidak ditemukan. Silakan ulangi program.")
            continue # melanjutkan ke input id barang
        stok_akhir = result[0] + int(barangIn) - int(barangOut) # menghitung stok akhir barang
        # mengeksekusi query update barang berdasarkan id barang
        crs.execute(f"UPDATE stok_barang SET stok_awal='{result[0]}', barang_masuk = '{barangIn}', barang_keluar = '{barangOut}', stok_akhir = '{stok_akhir}' WHERE id_barang = '{idbarang}'")
        db.commit() # menyimpan perubahan pada database
        print("Update Data Berhasil")
        print("-----------------------------------")
        pil = int(input("Update data lagi? 1.Ya \n 0. Tidak: ")) # menanyakan apakah ingin memngupdate data lagi atau tidak
```

```python
# membuat function untuk menghapus data barang
def hapus_data():
    pil = 1 # menginsialisasikan nilai awal variabel pil dengan 1
    while pil==1: # membuat perulangan while apabila pil==1 maka kode dibawah ini akan dijalankan 
        print("---------HAPUS DATA BARANG---------")
        crs.execute("SELECT * FROM stok_barang") # mengeksekui query untuk mengambil data barang
        result = crs.fetchall()
        for row in result: # menampilkan data barang menggunnakan perulanagan for
            idbarang = row[0]  # hanya menampilkan id dan nama barang saja
            nama = row[1]
            print("Data Barang")
            print(f"ID Barang: {idbarang}\nNama Barang: {nama}\n")

        idbarang_input = input("Masukkan ID Barang yang ingin dihapus: ") # meminta inputan id yang ingin dihapus
        # mengeksekusi query mengambil data dari table stok barang berdasarkan id barang
        crs.execute(f"SELECT stok_akhir FROM stok_barang WHERE id_barang = '{idbarang_input}'")
        res = crs.fetchone()
        if res is None: # membuat pengkondisian apabila id barang tidak ada maka akan menampilkna pesna error
            print(f"ID Barang {idbarang_input} tidak ditemukan")
            continue # lanjut dari menginput id barang
        # mengeksekusi query delete untuk menghpapus data barnag berdasarkan id barang
        crs.execute(f"DELETE FROM stok_barang WHERE id_barang = '{idbarang_input}'")
        db.commit() #menyimoan hasil perubahan pada database
        print("Hapus Data Berhasil")
        print("-----------------------------------")
        pil = int(input("Hapus data lagi? 1.Ya \n 0. Tidak: ")) # menanyakan apakah ingin menghapus data lagi atau tidak
```

```python
# mendefinisikan fungsi cari_data
def cari_data():
    pil = 1      # inisialisasi nilai pil dnegan 1
    while pil==1: # membuat perulangan while apabila pil==1 maka kode dibawah ini akan dijalankan 
        print("---------CARI DATA BARANG---------")
        keyword = input("Masukkan keywprd nama: ") # meminta input keyword dari user
        
        # mengeksekusi query untuk mencari data barang dengan nama yang mengandung keyword
        crs.execute(f"SELECT * FROM stok_barang WHERE nama_barang LIKE '%{keyword}%'")
        result = crs.fetchall() # mengambil hasil query
        if result:
            # jika data ditemukan
            for row in result: # menampikan data barang menggunakan perulangan for untuk menyimpan data berdasarkan row
                idbarang = row[0]
                nama = row[1]
                stok1 = row[2]
                barangIn = row[3]
                barangOut = row[4]
                stok2 = row[5]
            print("------------DATA BARANG-----------") # menampilkan data barang
            print(f"ID Barang: {idbarang}\nNama Barang: {nama}\
            \nStok Awal: {stok1}\nJumlah Barang Masuk: {barangIn}\nJumlah Barang Keluar: {barangOut}\nStok Akhir: {stok2}")
        else:
            # jika data tidak ditemukan
            print("Data tidak ditemukan") # menampilkan pesan error
        print("----------------------------------")
        # meminta input dari user untuk mencari data lagi atau tidak
        pil = int(input("Cari data lagi? 1.Ya \n 0. Tidak: "))
```
